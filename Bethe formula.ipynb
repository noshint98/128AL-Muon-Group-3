```python
import numpy as np


c=3e8
Beta=.1 #dummy variable amount, v/c
gamma=1/np.sqrt(1-Beta*Beta) #Lorentz factor
A=13 
Z=7 #atomic number of N
I=A*Z
z=1 #charge relative to electron charge
m=1.88e-28/1.6605e-27 #mass of muon in AU
p0=101325 #Pa
T0= 288.15 #K
g= 9.807 
L= .0065 #K/m
R= 8.31446 #J/mol/K
M= 0.0289652 #kg/mol
def get_C0(x,gamma,Beta):
    C0=0.3071*Z*z*z/(A*Beta*Beta)*(np.log(2*m*c*c*Beta*Beta*gamma*gamma/I)-Beta*Beta)
    return C0
def get_rho(x):
    rho=p0*M/(R*T0)*(1-L*x/T0)**(g*M/(R*L)-1)
    return rho
def get_t_prime(C0,rho,gamma1,gamma2):
    t_prime=m*c/(rho*C0)*(1/np.sqrt(gamma2**2-1)-1/np.sqrt(gamma1**2-1))*(gamma2-gamma1)
    return t_prime
    
x0=12000 #height of troposphere
deltax=1 #dummy amount, will change
E2=gamma*m*c**2 #double check, not correct
gamma2=gamma
Beta2=Beta
for i,x in enumerate(range(x0,0,-deltax)):
    C0=get_C0(x,gamma2,Beta2)
    rho=get_rho(x)
    dE=C0*rho*deltax 
    E1=E2
    E2=E1-dE
    gamma1=gamma2
    gamma2=E2*gamma1/E1
    Beta1=Beta2
    Beta2=np.sqrt(1-1/(gamma2*gamma2))

    t_prime=get_t_prime(C0,rho,gamma1,gamma2)
    if x%500==0:
        print(x,Beta2,E2,C0,rho,gamma2,t_prime,'\n')


    
    
```
